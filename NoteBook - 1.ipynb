```python
x = np.linspace(0,5,11)
y = x**2
```

*There are 2 methods to make a plot in matplotlib using 1) Functional 2) Object*

*FUNCTIONAL*

```python
plt.plot(x,y)
plt.xlabel('X-LABEL')
plt.ylabel('Y-LABEL')
plt.title('TITLE')
plt.show()   #Printing the plot, instead of just showing the plot
```

*MULTIPLE PLOTS ON THE SAME CANVAS BASICALLY WORKING WITH SUBPLOTS*

```python
plt.subplot(1,2,1)                   #plt.subplot(row,column,number_example)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

*Object oriented method*

```python
fig = plt.figure()   #This basically gives a blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8])        #making axes
axes.plot(x,y)          #plotting 
axes.set_xlabel('X - LABEL')
axes.set_ylabel('Y - LABEL')
axes.set_title('TITLE')
```

```python
fig = plt.figure()             # [left, bottom, width, height]
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.2, 0.3, 0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('INSIDE')
axes1.set_title('OUTSIDE')
```

*Subplots using object oriented programming*

```python
fig , axes = plt.subplots(nrows=1, ncols=2)

plt.tight_layout()                    #to make gap between figure


# for current_ax in axes:
#     current_ax.plot(x,y)
axes[0].set_title('FIRST PLOT')
axes[0].plot(x,y)
axes[0].set_xlabel('X0-LABEL')
axes[0].set_ylabel('Y0-LABEL')
axes[1].set_title('SECOND PLOT')
axes[1].plot(y,x)
axes[1].set_xlabel('X1-LABEL')
axes[1].set_ylabel('Y1-LABEL')


plt.tight_layout()                        #to take care of overlapping plots
```

```python
axes                   #axes object is just a list of all matplotlib axes in the subplots
```

*FIGURE SIZE, ASPECT RATIO AND DPI(DOTS PER INCH)*

```python
fig = plt.figure(figsize=(8, 2))                                #figsize in inches, you can change the figsize and play with it
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

*Can also do with subplots*

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=[8,2])
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
